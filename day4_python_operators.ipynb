## 1. Arithmetic Operators

Arithmetic operators are used to perform mathematical calculations such as addition, subtraction, multiplication, and division.

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

## Comparison Operators

Comparison operators are used to compare two values. The result of a comparison is always either True or False.

Common comparison operators:
- == : Equal to
- != : Not equal to
- > : Greater than
- < : Less than
- >= : Greater than or equal to
- <= : Less than or equal to

```python
x = 20
y = 15

print(x > y)
print(x < y)
print(x == y)
print(x != y)
```

## Assignment Operators

Assignment operators are used to assign values to variables and update their values.

Common assignment operators:
- = : Assign value
- += : Add and assign
- -= : Subtract and assign
- *= : Multiply and assign
- /= : Divide and assign
  








```python
num = 10

num += 5
print(num)

num -= 2
print(num)
```

## Logical Operators

Logical operators are used to combine multiple conditions and return a Boolean value (True or False).

Common logical operators:
- and : Returns True if both conditions are True
- or : Returns True if at least one condition is True
- not : Reverses the result

```python
age = 20

print(age > 18 and age < 30)
print(age > 25 or age < 30)
print(not(age > 18))
```

## Identity Operators

Identity operators are used to check whether two variables refer to the same object in memory.

Common identity operators:
- is : Returns True if both variables refer to the same object
- is not : Returns True if both variables refer to different objects

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)
print(a is c)
print(a is not c)
```

## Membership Operators

Membership operators are used to check whether a value exists in a sequence such as a list, tuple, string, or set.

Common membership operators:
- in : Returns True if the value exists
- not in : Returns True if the value does not exist

```python
fruits = ["apple",  "banana", "mango"]

print("apple" in fruits)
print("grapes" in fruits)
print("grapes" not in fruits)
print("mango" in fruits)
print("mango" not in fruits)
```

## Bitwise Operators

Bitwise operators perform operations on binary (bit-level) representations of numbers. They are commonly used in low-level programming and optimization.

Common bitwise operators:
- & : Bitwise AND
- | : Bitwise OR
- ^ : Bitwise XOR
- ~ : Bitwise NOT
- << : Left Shift
- >> : Right Shift

```python
a = 5
b = 3

print("AND:", a & b)
print("OR:", a | b)
print("XOR:", a ^ b)
print("NOT:", ~a)
print("Left Shift:", a << 1)
print("Right Shift:", a >> 1)
```

 ## Conclusion

Python operators are essential for performing calculations, making comparisons, combining conditions, assigning values, checking object identity, testing membership, and working with binary data. Understanding these operators is a fundamental step toward writing efficient and readable Python programs.
